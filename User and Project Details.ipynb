```python
import time
import random
import csv
from urllib.request import FancyURLopener
from random import choice
import bs4 as bs
from bs4 import SoupStrainer
from bs4 import BeautifulSoup

import pandas as pd

import json
```

```python
project = pd.read_csv('kickstarter_FRIDAY.csv')

project_id = project['project_id'].tolist()
project_url = project['project_url'].tolist()
```

```python
user_agents = [
    'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.36',
    'Opera/9.80 (X11; Linux i686; Ubuntu/14.10) Presto/2.12.388 Version/12.16',
    'Mozilla/5.0 (Windows; U; Windows NT 6.1; rv:2.2) Gecko/20110201',
    'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_9_3) AppleWebKit/537.75.14 (KHTML, like Gecko) Version/7.0.3 Safari/7046A194A',
    'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/42.0.2311.135 Safari/537.36 Edge/12.246'
]

class MyOpener(FancyURLopener, object):
    version = choice(user_agents)
    
```

```python
from bs4 import BeautifulSoup as bs
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
chrome_options = Options()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
```

```python
project_id_new = []
user_Facebook = []
user_BackingCount = []
user_TotalProject = []
user_Biography = []
project_description = []
project_risk = []
```

```python
for i in range(len(project_id)):
    
    myopener = MyOpener()
    
    url = project_url[i]
    
    if i % 100 == 0:
        print("project_id_new:- ", len(project_id_new))
        print("user_Facebook:- ", len(user_Facebook))
        print("user_BackingCount:- ", len(user_BackingCount))
        print("user_TotalProject:- ", len(user_TotalProject))
        print("user_Biography:- ", len(user_Biography))
        print("project_description:- ", len(project_description))
        print("project_risk:- ", len(project_risk))
        print("======>>>>><<<<<<=======")
    
    
    html_user = BeautifulSoup(myopener.open(url).read().decode('utf-8'), "html.parser")
    time.sleep(random.uniform(0,5))
    
    if html_user.find('div', class_ = 'bg-grey-100'):
    
        
        
        driver = webdriver.Chrome('/Users/dibshrestha/web Data/Codes/chromedriver', options = chrome_options)
        driver.get(url + '/description')
        time.sleep(random.uniform(2,5))
        html_project = driver.page_source
        driver.quit()



        #Get User Detail
        project_json = html_user.find('div', class_ = 'bg-grey-100').get('data-initial')
        project_json = json.loads(project_json)





        #Get Project Detail
        try:
            html_project = BeautifulSoup(html_project, "html.parser")
            story = html_project.find('section', class_="project-content")
            div = story.find('div', class_='rte__content')
            paragraph = div.find_all('p')
            text = ""
            for p in paragraph:
                text += p.get_text()
            risk = html_project.find('p', class_="js-risks-text text-preline").get_text()

            project_id_new.append(project_id[i])
            project_description.append(text)
            project_risk.append(risk)
            user_Facebook.append(project_json['project']['creator']['isFacebookConnected'])
            user_BackingCount.append(project_json['project']['creator']['backingsCount'])
            user_TotalProject.append(project_json['project']['creator']['launchedProjects']['totalCount'])
            user_Biography.append(project_json['project']['creator']['biography'])
        except:
            i = i - 2
            
            project_id_new = project_id_new[:i]
            user_Facebook = user_Facebook[:i]
            user_BackingCount = user_BackingCount[:i]
            user_TotalProject = user_TotalProject[:i]
            user_Biography = user_Biography[:i]
            project_description = project_description[:i]
            project_risk = project_risk[:i]
            
            print("Resting ====>>>")
            time.sleep(900)
            
            i -= 1
    
```

```python
outputdf = pd.DataFrame(list(zip(project_id_new, user_Facebook, user_BackingCount, user_TotalProject, user_Biography, project_description, project_risk)), 
               columns =['project_id', 'user_Facebook', 'user_BackingCount', 'user_TotalProject', 'user_Biography', 'project_description', 'project_risk'])
```

```python
outputdf.to_csv('kickstarter_details.csv', index=False)
```
