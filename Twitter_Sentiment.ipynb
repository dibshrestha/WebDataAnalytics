```python
from google.cloud import language
from google.cloud.language import enums
from google.cloud.language import types

import pandas as pd
import time
```

```python
import os
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = "web-data final-4346faceffdf.json"
```

```python
client = language.LanguageServiceClient()
```

```python
tweets = pd.read_csv('tweets_report.csv')
```

```python
tweets.head()
```

```python
tweet_uniqueProjectId = tweets.Project_ID.unique().tolist()

tweet_projectId = tweets.Project_ID.tolist()
tweet_text = tweets['Tweet Text'].tolist()

sentiment_list = []
magnitude_list = []
```

```python


i = 1

for text in tweet_text[]:
    
        
    if text != ' ':
        if i % 100 == 0:
            time.sleep(30)
            print(i)
        document = types.Document(
            content=text,
            type=enums.Document.Type.PLAIN_TEXT)
        sentiment = client.analyze_sentiment(document=document).document_sentiment
        sentiment_list.append(sentiment.score)
        magnitude_list.append(sentiment.magnitude)
        i+=1
    else:
        sentiment_list.append(0)
        magnitude_list.append(0)
    
    
```

```python
tweets['magnitude'] = magnitude_list
tweets['sentiment'] = sentiment_list
```

```python
followers = tweets['Followers'].tolist()
sentiment = tweets['sentiment'].tolist()
numerator = []
```

```python
for i in range(len(followers)):
    if followers[i] == ' ':
        numerator.append(0)
    else:
        numerator.append(float(followers[i]) * sentiment[i])
```

```python
tweets['numerator'] = numerator
```

```python
tweets.head(5)
```

```python
followers = []
len_tweet = []

for index, row in tweets.iterrows():
    if row['Followers'] == 0:
        followers.append(0)
        len_tweet.append(0)
    else:
        followers.append(int(row['Followers']))
        len_tweet.append(len(row['Tweet Text'] ))
 
        
tweets['Followers'] = followers
tweets['LenTweets'] = len_tweet
```

```python
sentiment_df = tweets.groupby(['Project_ID']).agg({'numerator':'sum','Followers':'sum', 'LenTweets':'sum'})
```

```python
sentiment_df['Sentiment_Result'] = sentiment_df['numerator']/sentiment_df['Followers']
```

```python
sentiment_df.fillna(0, inplace = True)
```

```python

sentiment_df.to_csv('twitter_sentiment.csv')
```
